```python
import os
```

```python
validation_data_dir = "./validation_data/Q&A"

tex_files = list(filter(lambda x: ".tex" in x, os.listdir(validation_data_dir)))

arxiv_id_to_lines = {}

for file in tex_files:
    arxiv_id = file[:-4]
    lines = []
    with open(os.path.join(validation_data_dir, file)) as f:
        for line in f:
            lines.append(line)
    arxiv_id_to_lines[arxiv_id] = lines

qna_files = list(filter(lambda x: ".json" in x, os.listdir(os.path.join(validation_data_dir))))

qnas = {}

for file in qna_files:
    arxiv_id = file[:-5]
    qnas[arxiv_id] = {}
    with open(os.path.join(validation_data_dir, file)) as f:
        qna = eval(f.read())
        for question, answers in qna.items():
            qnas[arxiv_id][question] = []
            for answer in answers:
                lines = arxiv_id_to_lines[arxiv_id]
                answer_text = " ".join(lines[answer[0]-1:answer[1]])
                qnas[arxiv_id][question].append(answer_text)
```

```python
import json

for arxiv_id, qna in qnas.items():
    print(arxiv_id, qna)
    with open(f"validation_data/{arxiv_id}.json", "w") as f:
        json.dump(qna, f)
```
